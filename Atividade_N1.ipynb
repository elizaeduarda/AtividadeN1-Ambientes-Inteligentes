### Questão 1

**A)** Descrição geral do dataset:

O dataset é composto por quatro arquivos principais, cada um fornecendo insights específicos sobre os filmes e as interações dos usuários.
Os arquivos principais que compõem o dataset são:

*   links.csv: Mapeia os filmes para seus identificadores no MovieLens, IMDb e TMDb.

**B)**  O que significa cada linha de cada arquivo?


No arquivo links.csv (df1), cada linha contém três informações principais: o ID do filme no MovieLens, o ID do filme no IMDb (que permite vincular o filme ao Internet Movie Database) e o ID do filme no TMDb (que conecta o filme ao The Movie Database).

```python
df1.head()
```



No arquivo movies.csv (df2), cada linha contém três informações principais: o ID do filme, o título do filme (incluindo o ano de lançamento entre parênteses) e os gêneros associados (separados por barras verticais).


```python
df2.head()
```

No arquivo ratings.csv (df3), cada linha contém quatro informações principais: o ID do usuário (userId), o ID do filme (movieId), a nota atribuída (rating, em uma escala de 0,5 a 5 estrelas) e o timestamp (que representa os segundos desde a meia-noite de 1º de janeiro de 1970, permitindo calcular a data exata em que a avaliação foi feita)

```python
df3.head()
```

No arquivo tags.csv (df4), cada linha contém quatro informações principais: o ID do usuário (userId), o ID do filme (movieId), a tag (uma palavra ou frase que o usuário associou ao filme) e o timestamp (que representa os segundos desde a meia-noite de 1º de janeiro de 1970, permitindo identificar a data exata em que a tag foi criada).

```python
df4.head()
```
